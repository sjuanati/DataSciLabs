```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Importing the dataset

```python
dataset = pd.read_csv('./filez/Position_Salaries.csv')

X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values

dataset
```

Some reminders:
- preprocessing categorical features:
  - order matters -> `LabelEncoder()`
  - order doesn't matter -> `OneHotEncoder()`
- feature scaling not needed for decision tree & random forest regressions: no equations, but algos looking for data ranges
- decision tree works better with many features, but not good when single or few ones

### Training the Decision Tree Regression model on the whole dataset

```python
from sklearn.tree import DecisionTreeRegressor

regressor = DecisionTreeRegressor(random_state=0)
regressor.fit(X, y)
```

### Predicting a new result

```python
# bad prediction: lower than it should be.
new_pred = regressor.predict([[6.5]])[0]
print(f'new prediction for 6.5 position: ${new_pred:,.2f}')
```

### Visualising the Decision Tree Regression results (higher resolution)

```python
# increase the density of the X points with step = 0.1 -> (1.1, 1.2, 1.3....)
X_grid = np.arange(np.amin(X), np.amax(X), 0.1)
X_grid = X_grid.reshape(len(X_grid), 1)
plt.scatter(X, y, color='red')
plt.plot(X_grid, regressor.predict(X_grid), color = 'blue')
plt.title('Decision Tree Regression')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

    -> Decision Tree Regression is not ideal for this dataset (considers that positions near an actual value are the same, which is not the case)
