```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Importing the dataset

```python
dataset = pd.read_csv('./filez/Position_Salaries.csv')

X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values

dataset
```

### Training the Random Forest Regression model on the whole dataset

```python
from sklearn.ensemble import RandomForestRegressor

regressor = RandomForestRegressor(n_estimators=10, random_state=0)
regressor.fit(X, y)
```

### Predicting a new result

```python
new_pred = regressor.predict([[6.5]])
print(f'new prediction for 6.5 position: ${new_pred[0]:,.2f}')
```

### Visualising the Random Forest Regression results (higher resolution)

```python
# increase the density of the X points with step = 0.1 -> (1.1, 1.2, 1.3....)
X_grid = np.arange(np.amin(X), np.amax(X), 0.1)
X_grid = X_grid.reshape(len(X_grid), 1)
plt.scatter(X, y, color='red')
plt.plot(X_grid, regressor.predict(X_grid), color = 'blue')
plt.title('Decision Tree Regression')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```
